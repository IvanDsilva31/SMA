```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded your movie dataset into a DataFrame named 'movie_data'
# If not, replace 'your_movie_dataset.csv' with your actual dataset file
movie_data = pd.read_csv('movies.csv')


```

```python
# Display pie chart of movie genres
genre_counts = movie_data['genre'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movie Genres')
plt.show()

```

```python
# Display bar chart of average score by year
average_score_by_year = movie_data.groupby('year')['score'].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.bar(average_score_by_year['year'], average_score_by_year['score'], color='skyblue')
plt.title('Average Movie Score by Year')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.grid(axis='y')
plt.show()
```
